#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
table= soup.find_all('table',{'class':'wikitable sortable'})[0]

languages= table.find_all('tr')
languages = [language.get_text().strip().split('\n\n') for language in languages]

colnames = languages[0]
data = languages[1:]

df = pd.DataFrame(data, columns=colnames)
df = df.astype({'Speakers(millions)': 'float'})
df.sort_values(by='Speakers(millions)', ascending=False,inplace=True)


df.head(10)
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
import re

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
info= soup.find_all('table',{'class':'chart full-width'})[0]



rows = info.find_all('tr')


#title
titles = [row.get_text().strip().split('\n')[1].strip() for row in rows]
titles = titles[1:]



#release year
years = [row.get_text().strip().split('\n')[2] for row in rows]
years = [re.sub(r'[()]', '', year) for year in years]
years = years[1:]



#rating
ratings = [row.get_text().strip().split('\n')[5] for row in rows[1:]]
ratings


a= info.find_all('a')
dirs = [row.get('title') for row in a]
dirs = dirs[1:]
dirs = [i for i in dirs if i] 

#actors
actors = [dir.split('(dir.)')[1] for dir in dirs]
actors = [re.sub('^[,]*', '', actor).strip() for actor in actors]

#director
dirs = [dir.split('(dir.)')[0] for dir in dirs]


rank = [i for i in range(1,251)]


dict = {'Rank':rank,'Title':titles,'Release Year':years,'Rating':ratings,'Director':dirs,'Actors':actors}

df = pd.DataFrame(dict)
df
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")


titles= soup.find_all('h3',{'class':'lister-item-header'})



titles = [title.get_text().strip().split('\n') for title in titles]


#titles
title = [title[1] for title in titles]

#year
year = [year[2] for year in titles]
year = [re.sub(r'[()]', '', y) for y in year]

#
summary = soup.find_all('p',{'class':''})
sum = [descrip.get_text().strip() for descrip in summary]
sum


dict = {'Title':title,'Year':year,'Summary':sum}

df = pd.DataFrame(dict)
df.head(10)

````
